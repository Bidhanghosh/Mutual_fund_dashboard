#### HISTORICAL_NAV

```python
import pandas as pd

listi = ["149389", "149072", "149463", "143247", "128627"]

df1 = pd.DataFrame()

for i in listi:
    # Get historical NAV data (assuming it's a list of dicts)
    historical_data = mf.get_scheme_historical_nav(i)
    
    # Convert the 'data' part to DataFrame; adjust key if needed
    nav_df = pd.DataFrame(historical_data['data'])
    
    # Extract scheme name
    name = mf.get_scheme_details(i)['scheme_name']
    
    # Add NAV column to df1, aligning by index (date)
    # Assuming nav_df has 'date' and 'nav' columns
    nav_df.set_index('date', inplace=True)
    df1[name] = nav_df['nav']

# If you want to see the combined DataFrame
print(df1)

```
